```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Read Data

```python
df = pd.read_csv('aws_faq.csv')
df.dropna(inplace = True)
```

```python
df
```

# Train the vectorizer

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(np.concatenate((df.Question, df.Answer)))
```

```python
vectorizer.vocabulary_
```

# Question Vectorizer

```python
Question_vectors = vectorizer.transform(df.Question)
```

# Chat With User

```python
print('Hey there I\'m available to talk !')
while True:
    input_question = input()
    input_question_vector = vectorizer.transform([input_question])
    similarity = cosine_similarity(input_question_vector, Question_vectors)
    closest = np.argmax(similarity, axis = 1)
    print('Bot:'+df.Answer.iloc[closest].values[0])
```
